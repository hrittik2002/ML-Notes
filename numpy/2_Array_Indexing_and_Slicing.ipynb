# 1. Numpy Indexing and Slicing of Vectors

## A. 1-D Array

````python
arr = np.arange(11,21)
arr
````

````python
## to print the index
arr[2]
````

````python
arr[1:5] # start index : end+1 index
````

````python
arr[:5] # default start value is 0
````

````python
arr[:] # default start value(0):default end value(size of arr)
````

````python
arr[3:] # default end (size of arr)
````

````python
arr[1::2] # the 3rd vaiable is step
# start index = 1
# end index = size of arr
# step = 2
````

## B. 2-D Array

````python
arr = np.arange(1,31).reshape(6,5)
arr
````

````python
arr[0] # print the 0th row
````

````python
arr[1][2] # print the 2nd element of 1st row
````

````python
arr[1,2] # same think diff syntax
````
