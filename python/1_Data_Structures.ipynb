## 7.B Array Methods

````python
# Append
l1 = ["Black" , "Green" , "Oolong", "White"]
l1.append("Lemon")
l1
````

````python
# Pop
l1 = ["Black" , "Green" , "Oolong", "White"]
l1.pop() # pop the last element and it returns the element removed
l1
````

````python
# remove
l1 = ["Black" , "Green" , "Oolong", "White"]
l1.remove("Green") # remove the element
l1
````

````python
# insert
l1 = ["Black" , "Green" , "Oolong", "White"]
l1.insert(1 , "Lemon") # insert at index 1
l1
````

````python
# copy
l1 = ["Black" , "Green" , "Oolong", "White"]
l2 = l1.copy()
print(l1)
print(l2)

l1[0] = "Lemon"
print(l1)
print(l2) # here l2 getting different reference so change in l1 not afftecting l2
````

````python
# copy vs =
l1 = ["Black" , "Green" , "Oolong", "White"]
l2 = l1
print(l1)
print(l2)

l1[0] = "Lemon"
print(l1)
print(l2) # here l2 getting same reference so change in l1 also afftecting l2
````

## 7.C List Comprehension


````python
# Normal way
squared_num = []
for x in range(10):
    squared_num.append(x**2)
squared_num
````

````python
# List Comprehension
squared_num = [x**2 for x in range(10)]
squared_num
````


Syntax
```
[expression   for variable in iterable]
    ↑                ↑            ↑
 what to do     loop variable    loop over what
```



# 8. Dictinary

- Key value pair
- mutable in nature
- unordered

````python
bio_data = {
    "name" : "Hrittik",
    "age" : "23",
    "country" : "India"
}
bio_data
````

````python
# how to get any value by key
bio_data["name"] , bio_data.get("name") # both ways works
````

````python
# bio_data["company"] # error , KeyError: 'company'
bio_data.get("company") # None
````

````python
# we can change any item
bio_data["name"] = "Hrittik Garain"
bio_data
````

````python
# printing dictionary
# Method 1
for key in bio_data:
  print(f"{key} : {bio_data[key]}")
````

````python
# Method 2
for key, value in bio_data.items():
  print(f"{key} : {value}")
````

````python
# length
len(bio_data)
````

````python
# adding new item
bio_data["fav_subject"] = "science"
bio_data
````

````python
# pop
bio_data.pop("fav_subject")
bio_data
````

````python
# copy
bio_data = {
    "name" : "Hrittik",
    "age" : "23",
    "country" : "India"
}
bio_data_copy = bio_data.copy()
print(bio_data)
print(bio_data_copy)

bio_data["name"] = "Hrittik Garain"
print(bio_data)
print(bio_data_copy)
````

````python
# dictionary comphrension
squared_num = {x : x**2 for x in range(5)}
squared_num
````

````python
# clear
squared_num.clear()
squared_num
````
