## 2.1. Types of Attributes

### a. Class Attribute
- Defined directly inside the class, outside any method. Shared across all objects.
- Class attributes are stored once in the class object in memory. All instances share the same reference — they don't get their own copy.

````python
class Car:
    wheels = 4  # class attribute — every Car has 4 wheels

car1 = Car()
car2 = Car()

print(car1.wheels)  # 4
print(car2.wheels)  # 4

# Change it on the class — affects ALL objects
Car.wheels = 6
print(car1.wheels)  # 6
print(car2.wheels)  # 6
````

### b. Instance Attribute
- Defined using self inside __init__. Each object has its own copy.
- Each instance attribute is stored in that object's own __dict__. So every time you create a new object, a new dictionary is allocated in memory holding only that object's instance attributes.

````python
class Car:
    wheels = 4          # class attribute — shared

    def __init__(self, color):
        self.color = color  # instance attribute — unique per

car1 = Car("Red")
car2 = Car("Blue")

print(car1.color)  # Red
print(car2.color)  # Blue  — completely independent
````

## 2.2. Types of Methods

### a. Instance Method
- Takes *self* as first argument
- Can read and modify instance attributes
- Called on an object



> self = the specific object that called the method. Python passes it automatically.



````python
class Car:
    def __init__(self, speed):
        self.speed = speed

    def accelerate(self, amount):   # instance method
        self.speed += amount
        print(f"Speed: {self.speed}")

car = Car(0)
car.accelerate(50)   # Speed: 50
````

### b. Class Method
- Takes ***cls*** as first argument (the class itself, not an object)
- Decorated with ***@classmethod***
- Can read and modify class attributes
- Cannot touch instance attributes — it has no self
- Can be called on the class directly — no object needed


> cls = the class itself. Just like self gives you the object, cls gives you the class.




````python
class Car:
    wheels = 4  # class attribute

    @classmethod
    def get_wheels(cls):
        return cls.wheels

    @classmethod
    def set_wheels(cls, count):
        cls.wheels = count

Car.set_wheels(6)
print(Car.get_wheels())  # 6
````

Most common real use — factory method (alternative constructor):

````python
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    @classmethod
    def create_default(cls):
        return cls("Toyota", "White")  # creates object from inside class

car = Car.create_default()
print(car.brand, car.color)  # Toyota White
````

### c. Static Method
- No self, no cls — takes neither object nor class
- Decorated with ***@staticmethod***
- Just a regular utility function that lives inside a class for logical grouping
- Cannot access or modify instance or class attributes directly



````python
class Car:
    @staticmethod
    def is_valid_speed(speed):
        return 0 <= speed <= 300

print(Car.is_valid_speed(150))  # True
print(Car.is_valid_speed(-10))  # False
````

When we use static methods?

- This logic is related to the class conceptually, but it doesn't need any data from the class or the object to do its job
- It's a helper/utility function that you group inside the class just because it logically belongs there.

Example:

You have a BankAccount class. Before creating an account, you want to validate the input. That validation doesn't need any object or class data — it just checks the input itself.

````python
class BankAccount:
    def __init__(self, owner, balance):
        if not BankAccount.is_valid_balance(balance):  # use static here
            raise ValueError("Invalid balance")
        self.owner = owner
        self.balance = balance

    @staticmethod
    def is_valid_balance(amount):       # no self, no cls needed
        return isinstance(amount, (int, float)) and amount >= 0

# Use before even creating an object
print(BankAccount.is_valid_balance(500))   # True
print(BankAccount.is_valid_balance(-100))  # False
````

# 3. Inheritance
- Inheritance allows a class (child class) to inherit properties and behaviors (attributes and methods) from another class (parent class)
- Why we need Inheritance:
  -  Code reusablity
  -  Organized structure
  -  Easy to maintain and extend

- Example : Lets create a Parent Class Car and a Child Class Electric Car. Here the child class will inherit all the properties from parent class like brand and model, and it will have some extra attributes like battery_capacity.

````python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def start_engine(self):
        print(f"The {self.brand} {self.model}'s engine is starting.")


class ElectricCar(Car): # parent class in parentheses
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity

my_tesla = ElectricCar("Tesla", "Model S", 75)

print(my_tesla.brand)
print(my_tesla.model)
print(my_tesla.battery_capacity)

my_tesla.start_engine()
````



-  In python, to create a child class we just need to put the parent class in parenthesis as above. No need to use any extend keyword like java or js



-  What is super() method ?
    - Lets say our child class (ElectricCar) needs new parameters (battery_capacity) then we have to create a constructor in child class. But the parameters(brand, model) that can be initialized in the parent class will be initialized using the Super() function.
    - targets the parent class.
    - super() works in any method, not just the constructor.
  

```python
class ElectricCar(Car):
    def start_engine(self):
        super().start_engine()        # runs parent version first
        print("Battery check done.")  # then adds extra behavior
```


- Lets say you have created a parent class with a constructor function inside it and then this class is inherited by a child class. Then the constructor function of parent class will work for the child class as well.





````python
class Animal:
    def __init__(self , name):
      self.name = name

    def show(self):
      print(f"Hello! your name is {self.name} ")


class Human(Animal):
  pass

human = Human("Hrittik") # we are using the constructor of parent class
human.show()
````

## 3.1. Types of Inheritance


### a. Single Inheritance
- One parent → One child
- Use when a class is a direct specialized version of one parent.


````python
class Animal:
    def breathe(self):
        print("Breathing...")

class Dog(Animal):      # Dog inherits from Animal
    def bark(self):
        print("Woof!")

dog = Dog()
dog.breathe()   # ✅ from Animal
dog.bark()      # ✅ from Dog
````

### b. Multiple Inheritance

- One child → Two or more parents.

````python
class Camera:
    def click_photo(self):
        print("Photo clicked")

class Phone:
    def make_call(self):
        print("Calling...")

class Smartphone(Camera, Phone):    # inherits from both
    def browse(self):
        print("Browsing internet")

s = Smartphone()
s.click_photo()     # from Camera Class
s.make_call()       # from Phone Class
s.browse()          # own method
````

#### The Diamond Problem — Comes With Multiple Inheritance
Look at this structure:


```
        A
       / \
      B   C
       \ /
        D
```
Both B and C inherit from A. Then D inherits from both B and C.

Now when D calls hello() — which one runs?

- B has hello()
- C has hello()
- A also has hello()

This ambiguity is called the Diamond Problem.

#### How Python Solves It — MRO
MRO = Method Resolution Order

It is the fixed sequence Python follows to look up any method or attribute.
Python uses an algorithm called C3 Linearization to calculate this order.

Simple rule to remember:

> Simple rule to remember:  
>  *Left to right through parents, depth first, but never visit a class before all its children are done.*

Example :



````python
class A:
    def hello(self):
        print("A")

class B(A):
    def hello(self):
        print("B")

class C(A):
    def hello(self):
        print("C")

class D(B, C):      # which hello() runs?
    pass

d = D()
d.hello()           # → B  (MRO: D → B → C → A)
print(D.__mro__)
````

Python solves this with MRO — always left to right through parents.

### c. Multilevel Inheritance

- Chain of inheritance. Grandparent → Parent → Child.

````python
class Animal:
    def breathe(self):
        print("Breathing...")

class Dog(Animal):          # Dog is child of Animal
    def bark(self):
        print("Woof!")

class Puppy(Dog):           # Puppy is child of Dog
    def play(self):
        print("Playing!")

puppy = Puppy()
puppy.breathe()     # ✅ from Animal (grandparent)
puppy.bark()        # ✅ from Dog (parent)
puppy.play()        # ✅ own method
````

# 4. Polymorphism
- Polymorphism is a core concept in OOPs. The word means "many forms" - and in programming, it allows the same interface or method name to behave differently depending on the object or context.
- Polymorphism can be achieved in python in 2 ways - well if we talk about compile time languages there are 3 ways - but python does not support  Method Overloading.
- Method Overriding means having same name methods inside a class but parameters will be different but in python the latest defination will overwrite the previous one.


## 4.1. Method Overriding
- This is where a child class overrides a method of the parent class and Python decides at runtime which method to call, based on the object type

- The child class always gets more priority.

````python
class Car:
    def start_engine(self):
        print("Generic engine starting")

class ElectricCar(Car):
    def start_engine(self):  # overrides parent
        print("Silent electric motor starting")

my_tesla = ElectricCar()
my_tesla.start_engine()  # → Silent electric motor starting
````

## 4.2. Duck Typing
- Duck typing allows you to write functions that are open to any object — as long as it has the required method. The benefit is that your function never needs to change when new types are added. You program to behavior, not to type

````python
class Dog:
    def speak(self):
        print("Woof!")

class Cat:
    def speak(self):
        print("Meow!")

class Robot:
    def speak(self):
        print("Beep boop!")

def make_it_speak(anything):   # no type check — just calls speak()
    anything.speak()

make_it_speak(Dog())    # Woof!
make_it_speak(Cat())    # Meow!
make_it_speak(Robot())  # Beep boop!
````

Dog, Cat, Robot have zero relationship to each other. No common parent. No interface. No inheritance.

But the function doesn't care. It just asks — "do you have a speak() method?" If yes — it works.

```python
        anything.robot_speak()
    # every new class = come back and edit this function
```

# 5. Encapsulation

Encapsulation means buinding all the data and methods together inside a class and controlling who can access or modify the data.  
In Simple term:

*   buinding all the data and methods together : Before OOPs all the data and methods use to be scattered. But now we can put everything togerther inside a class.
*   controlling who can access or modify the data : We make the class data private and use setter and getter methods to modify or access the data.



## 5.1. Access Modifiers

Access modifiers control who can access attributes and methods. Python has 3 levels — but unlike Java or C++, Python does not enforce them at compile time. They work through naming conventions and name mangling.

````python
class Car:
    def __init__(self, brand, speed, engine_code):
        self.brand = brand                    # public    → anyone can access
        self._speed = speed                   # protected → "internal use" (convention only)
        self.__engine_code = engine_code      # private   → name-mangled by Python

car = Car("Toyota", 120, "V8-SECRET")
````

### a. Public
- No underscore prefix.
- Accessible from anywhere — objects, child classes, outside modules.


```python
print(car.brand)   # ✅ Toyota — works fine
```


### b. Protected
- Single underscore prefix: `_speed`
- Convention only — Python does NOT block access. It's a signal to developers: "this is for internal use, don't touch it from outside unless you're a subclass."
- Java enforces protected at compile time. Python does not.


```python
print(car._speed)  # ✅ 120 — Python allows it, but you shouldn't do this outside the class

# Child class accessing protected attribute — this is the intended use
class ElectricCar(Car):
    def show_speed(self):
        print(self._speed)   # ✅ correct usage inside child class
```



### c. Private
- Double underscore prefix: `__engine_code`
- Python applies **name mangling** — it renames the attribute internally to `_ClassName__attribute`.
- This prevents accidental access but is NOT true security. You can still access it if you know the mangled name.

```python
# Direct access fails
# car.__engine_code   ❌ AttributeError

# Name mangling — Python stores it as _Car__engine_code
print(car._Car__engine_code)   # ✅ V8-SECRET  (works but you shouldn't do this)

# Check what Python actually stored
print(car.__dict__)
# {'brand': 'Toyota', '_speed': 120, '_Car__engine_code': 'V8-SECRET'}
```





## 5.2. Getters and Setters

Since we make data private, we need controlled ways to read and modify it. There are 2 approaches in Python.

### Approach 1 — Traditional getter/setter methods
Like Java style. Works fine but not very Pythonic.

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance       # private

    # Getter — read the balance
    def get_balance(self):
        return self.__balance

    # Setter — update with validation
    def set_balance(self, amount):
        if amount < 0:
            raise ValueError("Balance cannot be negative")
        self.__balance = amount

account = BankAccount("Hrittik", 5000)

print(account.get_balance())     # 5000

account.set_balance(8000)
print(account.get_balance())     # 8000

# account.set_balance(-500)      # ❌ ValueError: Balance cannot be negative
````

### Approach 2 — @property (Pythonic way — preferred)

`@property` lets you write getter and setter methods but call them like normal attributes. Clean syntax, same protection.

Why it's better:
- Looks cleaner: `account.balance` instead of `account.get_balance()`
- Adds validation on write without changing the interface
- You can start with a public attribute and add a property later — callers don't need to change their code

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance       # private — store with __

    @property
    def balance(self):                 # getter — called when you READ balance
        return self.__balance

    @balance.setter
    def balance(self, amount):         # setter — called when you WRITE balance
        if amount < 0:
            raise ValueError("Balance cannot be negative")
        self.__balance = amount

account = BankAccount("Hrittik", 5000)

# Reads like a normal attribute — but actually calls the getter
print(account.balance)       # 5000

# Writes like a normal attribute — but actually calls the setter with validation
account.balance = 8000
print(account.balance)       # 8000

# account.balance = -500     # ❌ ValueError: Balance cannot be negative
````

If you define `@property` but no `@x.setter`, the attribute becomes read-only.

# 6. Abstraction
- Abstraction does not exist in python but we can achieve it using a library we will see what is a library later.
- Abstraction means simplifying complex systems by focusing on essential features and hiding unnecessary details.
- It is used to define a common interface for different subclasses


## 6.1. Abstract Classes and Methods
- Abstract Classes are classes that contains one or more abstract methods.
- A method that is defined but not implemented in the abstract class, subclasses must provide the implementation.

````python
from abc import ABC, abstractmethod

class Shape(ABC):

  @abstractmethod
  def area(self):
    pass

  @abstractmethod
  def perimeter(self):
    pass

# Rectangle Class is implementing the abstract methods of Shape
class Rectangle(Shape):
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
    return self.length * self.width

  def perimeter(self):
    return 2 * (self.length + self.width)


# Circle Class is implementing the abstract methods of Shape
class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return 3.14 * self.radius * self.radius

  def perimeter(self):
    return 2 * 3.14 * self.radius

rect = Rectangle(5, 4)
print(rect.area())
print(rect.perimeter())
circle = Circle(7)
print(circle.area())
print(circle.perimeter())
````

# 7. Dunder Methods
- Dunder means double underscore, the __ before and after the method name.
- example :  

```
__init__      # dunder init
__str__       # dunder str
__len__       # dunder len
```


- This are called magic methods.
- Dunder methods let your custom class behave like a built-in Python type. Dunder methods are the bridge between your class and Python's built-in behavior.

## __ init __ — Constructor


````python
class Car:
    def __init__(self, brand):   # called when object is created
        self.brand = brand

car = Car("Toyota")   # triggers __init__ automatically
````

## __ str __ — Human Readable Output

Called when you print() an object.

````python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def __str__(self):
        return f"{self.brand} going at {self.speed}kmph"

car = Car("Toyota", 120)
print(car)
# With __str__
# Toyota going at 120kmph
# Without __str__
# <__main__.Car object at 0x7f3c2d1a3d90>  ← ugly memory address
````

## __ repr __ — Developer/Debug Output
- Called in console, logs, debugging.
- Rule of thumb — __ str __ for users, __ repr __ for developers.


````python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def __repr__(self):
        return f"Car(brand='{self.brand}', speed={self.speed})"

car = Car("Toyota", 120)
print(repr(car))   # Car(brand='Toyota', speed=120)
````

## __ len __ — len() on Your Object



````python
class Garage:
    def __init__(self, cars):
        self.cars = cars

    def __len__(self):
        return len(self.cars)

garage = Garage(["Toyota", "Tesla", "BMW"])
print(len(garage))    # 3
````

## __ add __ — + Operator on Your Object


````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2          # triggers __add__ automatically
print(v3)             # Vector(4, 6)
````

## __ eq __ — == Comparison


````python
class Car:
    def __init__(self, brand):
        self.brand = brand

    def __eq__(self, other):
        return self.brand == other.brand

car1 = Car("Toyota")
car2 = Car("Toyota")
car3 = Car("BMW")

print(car1 == car2)   # True  — same brand
print(car1 == car3)   # False — different brand
# Note without __eq__ : False ← compares memory addresses, not values
````
